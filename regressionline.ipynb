```python
import matplotlib.pyplot as mpl
import csv
import numpy as np
import pandas as pd

# df = pd.read_csv('computingdata.csv')

#df['Absorbance'] = pd.to_numeric(df['Absorbance'], errors='coerce')

#df = df.dropna(subset=['Absorbance'])

#Time = df['Time'].tolist()
#Absorbance = df['Absorbance'].tolist()

Time = []
Absorbance = []

with open ('computingdata.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    throwaway = lines.__next__()
    for row in lines:
        Time.append(row[0])
        Absorbance.append(row[1])

mpl.scatter(Time, Absorbance, color = 'b')
mpl.xticks(rotation = 25)
mpl.xlabel('Time (s)')
mpl.ylabel('Absorbance (nm)')
mpl.title('Absorbance vs Time Graph', fontsize = 20)

# polyfit function from Numpy fits a linear line to the data points with a degree of 1
# polyl1d used to make a polynomial function from coefficients
z = np.polyfit(np.float32(Time), np.float32(Absorbance), 1)
p = np.poly1d(z)

# regression line plotted:
mpl.plot(Time, p(np.float32(Time)), "r--")

mpl.show()
```

```python
import matplotlib.pyplot as mpl
import csv
import numpy as np

Time = []
Absorbance = []

with open ('computingdata.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    throwaway = lines.__next__()
    for row in lines:
        Time.append(float(row[0]))
        Absorbance.append(float(row[1]))

mpl.scatter(Time, Absorbance, color = 'b')
mpl.xticks(rotation = 25)
mpl.xlabel('Time (s)')
mpl.ylabel('Log Absorbance (nm)')
mpl.title('Log Absorbance vs Time Graph', fontsize = 20)

#transformed data for 
z = np.polyfit(np.float32(Time), np.log(np.float32(Absorbance)), 1)
p = np.poly1d(z)
mpl.plot(Time, np.exp(p(np.float32(Time))), "r--")

mpl.show()
```
